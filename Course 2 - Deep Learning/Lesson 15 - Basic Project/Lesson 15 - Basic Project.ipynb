```python
from fpdf import FPDF

def create_pdf(result_dict, file_name):
    pdf = FPDF()
    pdf.add_page()

    pdf.set_font("Arial", size = 12)

    pdf.cell(200, 10, txt="Result Details", ln=True, align='C')

    pdf.cell(200, 10, txt=f"Score: {result_dict['score']}", ln=True)
    pdf.cell(200, 10, txt=f"Start Position: {result_dict['start']}", ln=True)
    pdf.cell(200, 10, txt=f"End Position: {result_dict['end']}", ln=True)
    pdf.cell(200, 10, txt=f"Answer: {result_dict['answer']}", ln=True)

    pdf_output_path = '/content/drive/My Drive/'+file_name+'.pdf'
    pdf.output(pdf_output_path)

    return pdf_output_path
```

```python
def create_pdf_for_summarization(text, file_name):
    pdf = FPDF()
    pdf.add_page()

    pdf.set_font("Arial", size=12)

    pdf.cell(200, 10, txt="Result Details", ln=True, align='C')

    text_lines = text.split('\n')
    for line in text_lines:
        pdf.multi_cell(0, 10, txt=line)

    pdf_output_path = '/content/drive/My Drive/' + file_name + '.pdf'
    pdf.output(pdf_output_path)

    return pdf_output_path
```

#Summarizer

```python
from transformers import pipeline

summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
# @title Summarizer
texto_a_resumir = "DevOps es una metodología que busca integrar de manera efectiva el desarrollo de software (Dev) con las operaciones de sistemas (Ops), facilitando así la entrega continua y la mejora constante en el ciclo de vida de desarrollo de aplicaciones. Este enfoque promueve la colaboración estrecha entre equipos de desarrollo, operaciones y calidad, utilizando herramientas y prácticas automatizadas para acelerar la entrega de software de alta calidad al cliente final. DevOps fomenta una cultura de responsabilidad compartida, transparencia y aprendizaje continuo, con el objetivo de optimizar la eficiencia, la fiabilidad y la escalabilidad de los sistemas de software." # @param {type:"string"}
```

```python
ARTICLE = pipe(texto_a_resumir)
```

```python
ARTICLE
```

```python
result_summarizer = summarizer(ARTICLE[0]['translation_text'], max_length=130, min_length=50, do_sample=False)
```

```python
result_summarizer
```

```python
src_text = [
    ">>esp<< "+result_summarizer[0]['summary_text'],
]

model = MarianMTModel.from_pretrained(model_name)
translated = model.generate(**tokenizer(src_text, return_tensors="pt", padding=True))
value = [tokenizer.decode(t, skip_special_tokens=True) for t in translated]
```

```python
pdf_path = create_pdf_for_summarization(value[0],'resultado_summarization')
print(f"PDF guardado en: {pdf_path}")
```
