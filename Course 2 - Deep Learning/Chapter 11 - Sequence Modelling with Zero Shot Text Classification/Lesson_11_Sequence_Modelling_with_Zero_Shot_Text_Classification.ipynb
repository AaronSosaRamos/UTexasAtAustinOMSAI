#Sequence Modelling with Zero Shot Text Classification

Source: https://huggingface.co/facebook/bart-large-mnli

```python
from transformers import pipeline
classifier = pipeline("zero-shot-classification",
                      model="facebook/bart-large-mnli")
```

```python
sequence_to_classify = "one day I will see the world"
candidate_labels = ['travel', 'cooking', 'dancing']
classifier(sequence_to_classify, candidate_labels)
```

Source: https://huggingface.co/MoritzLaurer/bge-m3-zeroshot-v2.0

```python
from transformers import pipeline
text = "Angela Merkel is a politician in Germany and leader of the CDU"
hypothesis_template = "This text is about {}"
classes_verbalized = ["politics", "economy", "entertainment", "environment"]
zeroshot_classifier = pipeline("zero-shot-classification", model="MoritzLaurer/deberta-v3-large-zeroshot-v2.0")  # change the model identifier here
output = zeroshot_classifier(text, classes_verbalized, hypothesis_template=hypothesis_template, multi_label=False)
print(output)
```

Source: https://huggingface.co/Recognai/bert-base-spanish-wwm-cased-xnli

```python
from transformers import pipeline
classifier = pipeline("zero-shot-classification",
                       model="Recognai/bert-base-spanish-wwm-cased-xnli")

classifier(
    "El autor se perfila, a los 50 años de su muerte, como uno de los grandes de su siglo",
    candidate_labels=["cultura", "sociedad", "economia", "salud", "deportes"],
    hypothesis_template="Este ejemplo es {}."
)
```

Source: https://huggingface.co/morit/spanish_xlm_xnli

```python
from transformers import pipeline
classifier = pipeline("zero-shot-classification",
                      model="morit/spanish_xlm_xnli")
sequence_to_classify = "Creo que Lionel Messi es el mejor futbolista del mundo."


# we can specify candidate labels and hypothesis:
candidate_labels = ["politíca", "futbol"]
hypothesis_template = "Este ejemplo es {}"

# classify using the information provided
classifier(sequence_to_classify, candidate_labels, hypothesis_template=hypothesis_template)
```
