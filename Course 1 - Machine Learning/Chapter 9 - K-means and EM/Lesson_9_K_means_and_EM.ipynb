#K-means Clustering
K-means is a popular clustering algorithm that partitions data into K clusters by iteratively assigning data points to the nearest cluster center and then updating the cluster centers based on the mean of the assigned points.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Apply K-means clustering
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Plotting the clusters
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', alpha=0.7)
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.9, marker='s')
plt.title('K-means Clustering')
plt.show()
```

#Elbow Method for Optimal *k*

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Try different values of k and calculate inertia (within-cluster sum of squares)
inertias = []
k_values = range(1, 11)  # Testing k from 1 to 10

for k in k_values:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

# Plotting the elbow curve
plt.plot(k_values, inertias, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Within-cluster sum of squares (inertia)')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.cm as cm

# Generate synthetic data
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Try different values of k and calculate silhouette scores
k_values = range(2, 11)  # Testing k from 2 to 10

for k in k_values:
    kmeans = KMeans(n_clusters=k)
    labels = kmeans.fit_predict(X)
    silhouette_avg = silhouette_score(X, labels)
    sample_silhouette_values = silhouette_samples(X, labels)

    plt.figure(figsize=(8, 6))
    plt.plot()

    y_lower = 10
    for i in range(k):
        ith_cluster_silhouette_values = sample_silhouette_values[labels == i]
        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / k)
        plt.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        plt.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
        y_lower = y_upper + 10

    plt.title("Silhouette plot for {} clusters".format(k))
    plt.xlabel("Silhouette coefficient values")
    plt.ylabel("Cluster label")
    plt.axvline(x=silhouette_avg, color="red", linestyle="--")
    plt.yticks([])  # Clear the yaxis labels / ticks
    plt.show()
```

#Expectation-Maximization (EM) Clustering (Gaussian Mixture Model)


EM clustering is a soft clustering algorithm based on Gaussian Mixture Models (GMM), where each cluster is modeled as a Gaussian distribution. It iteratively estimates the parameters of these distributions to maximize the likelihood of the observed data.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.mixture import GaussianMixture

# Generate moon-shaped synthetic data
Xmoon, _ = make_moons(200, noise=.05, random_state=0)

# Apply Gaussian Mixture Model (EM clustering)
gmm = GaussianMixture(n_components=2)
gmm.fit(Xmoon)
labels = gmm.predict(Xmoon)

# Plotting the clusters
plt.scatter(Xmoon[:, 0], Xmoon[:, 1], c=labels, cmap='viridis', alpha=0.7)
plt.title('EM Clustering (GMM)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.mixture import GaussianMixture

# Generate synthetic data with 4 clusters
X, _ = make_blobs(n_samples=400, centers=4, cluster_std=0.60, random_state=0)

# Initialize a Gaussian Mixture Model with 4 components
gmm = GaussianMixture(n_components=4, random_state=0)

# Fit the model to the data
gmm.fit(X)

# Predict cluster labels
labels = gmm.predict(X)

# Plotting the clusters and GMM contours
plt.figure(figsize=(12, 8))

# Scatter plot for data points colored by cluster
plt.subplot(121)
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', alpha=0.7)
plt.colorbar(label='Cluster')
plt.title('EM Clustering using Gaussian Mixture Model')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Contour plot for Gaussian components
plt.subplot(122)
x = np.linspace(-2, 8)
y = np.linspace(-2, 8)
X_grid, Y_grid = np.meshgrid(x, y)
XX = np.array([X_grid.ravel(), Y_grid.ravel()]).T
Z = -gmm.score_samples(XX)
Z = Z.reshape(X_grid.shape)
plt.contour(X_grid, Y_grid, Z, levels=10, cmap='viridis')
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', alpha=0.7)
plt.colorbar(label='Cluster')
plt.title('GMM Contours and Cluster Points')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.tight_layout()
plt.show()

# Print model parameters
print("Gaussian Mixture Model:")
print("Number of components:", gmm.n_components)
print("Converged:", gmm.converged_)
print("Weights of each component:", gmm.weights_)
print("Means of each component:\n", gmm.means_)
print("Covariance matrices of each component:\n", gmm.covariances_)
```
